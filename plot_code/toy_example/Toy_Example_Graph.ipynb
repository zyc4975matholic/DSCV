```python
# This code is implemented all by Siyu Gao. The code generates the toy-example demonstrated in both presentation and paper.
# The code regenerates the figures with the npy file in the directory.
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
class1_1 = np.load('class1_1.npy')
class2_1 = np.load('class2_1.npy')
class3_1 = np.load('class3_1.npy')
display(len(class2_1))
class1_2 = np.load('class1_2.npy')
class2_2 = np.load('class2_2.npy')
class3_2 = np.load('class3_2.npy')
display(len(class2_2))
class1_3 = np.load('class1_3.npy')
class2_3 = np.load('class2_3.npy')
class3_3 = np.load('class3_3.npy')

class1_4 = np.load('class1_4.npy')
class2_4 = np.load('class2_4.npy')
class3_4 = np.load('class3_4.npy')
class4_4 = np.load('class4_4.npy')
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 4,figsize = (25,5))
plt.subplots_adjust(wspace =0.3, hspace =0)
sns.set_style('white')

axes[0].scatter(x = class1_1[:,0],y = class1_1[:,1],label = 'known class 1',alpha = 0.5)
axes[0].scatter(x = class2_1[:,0],y = class2_1[:,1],label = 'known class 2')
axes[0].scatter(x = class3_1[:,0],y = class3_1[:,1],label = 'unknown class')
axes[0].legend()

axes[1].scatter(x = class1_2[:,0],y = class1_2[:,1],label = 'known class 1')
axes[1].scatter(x = class2_2[:,0],y = class2_2[:,1],label = 'known class 2')
axes[1].scatter(x = class3_2[:,0],y = class3_2[:,1],label = 'unknown class')

axes[2].scatter(x = class1_3[:,0],y = class1_3[:,1],label = 'known class 1')
axes[2].scatter(x = class2_3[:,0],y = class2_3[:,1],label = 'known class 2')
axes[2].scatter(x = class3_3[:,0],y = class3_3[:,1],label = 'unknown class',color = 'g',edgecolors='black',)

axes[3].scatter(x = class1_4[:,0],y = class1_4[:,1],label = 'known class 1')
axes[3].scatter(x = class2_4[:,0],y = class2_4[:,1],label = 'known class 2')
axes[3].scatter(x = class3_4[:,0],y = class3_4[:,1],label = 'unknown class')
# axes[3].scatter(x = class4_4[:,0],y = class4_4[:,1],label = 'predict unknown class')
```
