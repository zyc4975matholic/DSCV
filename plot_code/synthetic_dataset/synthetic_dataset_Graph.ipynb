```python
# This code is implemented all by Siyu Gao. This code generates the figures of DSCV versus class separability and DSCV versus imbalanced ratio on synthetic datasets.
# This code regenerates the figures with the txt file in this directory
import seaborn as sns
import pandas as pd
import numpy as np
import xlrd
import matplotlib.pyplot as plt
```

```python
df_1 = pd.read_table('section1_sep=1.txt')
df_4 = pd.read_table('section1_sep=4.txt')
df_7= pd.read_table('section1_sep=7.txt')
df_10 = pd.read_table('section1_sep=10.txt')
name_list = ['IR','F_1 benchmark','F_1 improved','GM benchmark','GM improved']
score = ['F_1 benchmark','F_1 improved','GM benchmark','GM improved']
df_1.columns = name_list
df_4.columns = name_list
df_7.columns = name_list
df_10.columns = name_list
```

```python
fig,axes = plt.subplots(figsize = (10,8))

index_1 = []
for i in range(0,40,2):
    index_1.append(i)

sns.lineplot(x = [df_1['IR'][i] for i in index_1],y = [df_1[score[2]][i] for i in index_1], label = 'ill-sep + SVM', color = 'lightgreen')
sns.lineplot(x =[df_1['IR'][i] for i in index_1],y = [df_1[score[3]][i] for i in index_1], label = 'ill-sep + DSCV (*)', color = 'green')
sns.lineplot(x =[df_1['IR'][i] for i in index_1],y = [df_10[score[2]][i] for i in index_1], label = 'well-sep + SVM',color = 'skyblue')
sns.lineplot(x =[df_1['IR'][i] for i in index_1],y = [df_10[score[3]][i] for i in index_1], label = 'well-sep + DSCV (o)',color = 'steelblue')

axes.lines[0].set_linestyle("--")
axes.lines[2].set_linestyle("--")

index = []
for i in range(0,40,4):
    index.append(i)
x = [df_1['IR'][i] for i in index]
y_1 = [df_1[score[3]][i] for i in index]
y_10 = [df_10[score[3]][i] for i in index]
plt.plot(x, y_1, '*')
plt.plot(x ,y_10, 'o')
axes.legend(loc = 'upper right',prop={'size': 12})
# plt.suptitle('DSCV Performance v.s Imbalance Ratio',fontsize=12)

# plt.legend(loc = 'lower right',prop={'size': 12})
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.xlabel(r'Imbalanced Ratio ($\rho$)',fontsize = 14)
plt.ylabel('G mean Score',fontsize = 14)
fig.savefig('pic_for_pre_1',bbox_inches = 'tight',dpi = 100)
```

```python
dfr_1 = pd.read_table('section2_ratio=1.txt')
dfr_50 = pd.read_table('section2_ratio=50.txt')
dfr_100 = pd.read_table('section2_ratio=100.txt')
dfr_200 = pd.read_table('section2_ratio=200.txt')
name_list = ['IR','sep_1','sep_2','F_1 benchmark','F_1 improved','GM benchmark','GM improved']
score = ['F_1 benchmark','F_1 improved','GM benchmark','GM improved']
dfr_1.columns = name_list
dfr_50.columns = name_list
dfr_100.columns = name_list
dfr_200.columns = name_list
```

```python
fig_1 = plt.figure(figsize = (10,8))

sns.lineplot(x = 'sep_1',y = score[3] ,data = dfr_1, color = 'pink',label = r'$\rho$ = 1 ')
sns.lineplot(x = 'sep_1',y = score[3] ,data = dfr_50,color = 'hotpink',label = r'$\rho$ = 50 ')
sns.lineplot(x = 'sep_1',y = score[3] ,data = dfr_100,color = 'orange',label = r'$\rho$ = 100 ')
sns.lineplot(x = 'sep_1',y = score[3] ,data = dfr_200,color = 'crimson',label = r'$\rho$ = 200 ')

plt.legend(loc = 'lower right',prop={'size': 12})
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.xlabel(r'$J_1$ score',fontsize = 14)
plt.ylabel('G mean score',fontsize = 14)
# plt.title('DSCV Performance v.s class separability')
fig_1.savefig('pic_for_pre_2',bbox_inches = 'tight',dpi = 100)
```
