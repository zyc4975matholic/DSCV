```python
# This is code is implemented all by Siyu Gao. This code generates the figures for visual illustration of ill-separated and well-separated data
# This code regenerates the figures with the npy file in the directory
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
class1_1i = np.load('class1_1(1).npy')
class2_1i = np.load('class2_1(1).npy')
class3_1i = np.load('class3_1(1).npy')
class1_1w = np.load('class1_1(2).npy')
class2_1w= np.load('class2_1(2).npy')
class3_1w = np.load('class3_1(2).npy')
```

```python
fig = plt.figure(figsize = (4,4))
sns.set_style('white')

plt.scatter(x = class1_1i[:,0],y = class1_1i[:,1],label = 'class 1',alpha = 0.8)
plt.scatter(x = class2_1i[:,0],y = class2_1i[:,1],label = 'class 2',alpha = 0.8,marker = 'x')
plt.scatter(x = class3_1i[:,0],y = class3_1i[:,1],label = 'class 3',alpha = 0.8,marker = '*')
plt.legend()
plt.xlabel('feature x')
plt.ylabel('feature y')
fig.savefig('ill_sep')
```

```python
fig = plt.figure(figsize = (4,4))
sns.set_style('white')

plt.scatter(x = class1_1w[:,0],y = class1_1w[:,1],label = 'class 1',alpha = 0.8)
plt.scatter(x = class2_1w[:,0],y = class2_1w[:,1],label = 'class 2',alpha = 0.8,marker = 'x')
plt.scatter(x = class3_1w[:,0],y = class3_1w[:,1],label = 'class 3',alpha = 0.8,marker = '*')
plt.legend()
plt.xlabel('feature x')
plt.ylabel('feature y')
fig.savefig('well_sep')
```
